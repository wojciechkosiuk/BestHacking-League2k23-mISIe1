```python
import pandas as pd
import nltk
import IProgress
import ipywidgets
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from tqdm.notebook import tqdm
tqdm.pandas()

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Define the preprocessing function
def preprocess_text(text):
    # Tokenize the text
    tokens = word_tokenize(str(text).lower())
    # Remove stop words
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]
    # Lemmatize the words
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    # Join the tokens back into a string
    #preprocessed_text = ' '.join(tokens)
    return tokens

# Apply the preprocessing function to the 'Description' column
df2['Preprocessed Description'] = df2['Description'].progress_apply(preprocess_text)

df2
```

```python
def clean_text(text_vec):
    return ' '.join(text_vec)

df2['clean_text'] = df2['Preprocessed Description'].progress_apply(clean_text)
df2
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# Create a bag of words from the preprocessed text
vectorizer = CountVectorizer()
bag_of_words = vectorizer.fit_transform(df2['clean_text'])
```

```python
# Get a list of the words in the bag of words
words = vectorizer.get_feature_names_out()

# Get the number of appearances of each word
word_counts = bag_of_words.sum(axis=0)

# Create a DataFrame with the word counts
word_counts_df = pd.DataFrame({'word': words, 'count': word_counts.tolist()[0]})
```

```python
word_counts_df.sort_values(by=['count'],ascending=False).head(20)
```

```python
import re
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple', 'pink', 'brown', 'black', 'white', 'gray']

# Define the function to extract color from preprocessed description
def extract_color(description):
    # Search for colors in the description
    matches = re.findall('|'.join(colors), description, flags=re.IGNORECASE)
    # If no color is found, return None
    if len(matches) == 0:
        return None
    # If multiple colors are found, return the first one
    else:
        return matches[0]

# Apply the function to the 'Preprocessed Description' column to create a new 'Color' column
df2['Color'] = df2['clean_text'].progress_apply(extract_color)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create a TF-IDF vectorizer object
vectorizer = TfidfVectorizer()

# Fit the vectorizer to the cleaned descriptions
vectorizer.fit(df2['clean_text'])

# Transform the cleaned descriptions into TF-IDF vectors
vectors = vectorizer.transform(df2['clean_text'])
```

```python
import spacy

# Load the spacy model
nlp = spacy.load('en_core_web_md')

# Function to get the Word2Vec vector for a list of words
def get_vector(full_text):
    # Initialize an empty vector
    doc = nlp(full_text)

    return doc.vector

# Apply the get_vector function to the cleaned descriptions
df2['word2vec_vector'] = df2['clean_text'].progress_apply(get_vector)
df2
# Print the resulting matrix of vectors
#vectors = np.stack(df2['word2vec_vector'].to_numpy())
#print(vectors)
```
