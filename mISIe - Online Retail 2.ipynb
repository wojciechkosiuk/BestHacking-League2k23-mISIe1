```python
import pandas as pd
import nltk
#import IProgress
import ipywidgets
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from tqdm.notebook import tqdm
tqdm.pandas()

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Define the preprocessing function
def preprocess_text(text):
    # Tokenize the text
    tokens = word_tokenize(str(text).lower())
    # Remove stop words
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]
    # Lemmatize the words
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    # Join the tokens back into a string
    #preprocessed_text = ' '.join(tokens)
    return tokens

# Apply the preprocessing function to the 'Description' column
df3['Preprocessed Description'] = df3['Description'].progress_apply(preprocess_text)

df3
```

```python
def clean_text(text_vec):
    return ' '.join(text_vec)

df3['clean_text'] = df3['Preprocessed Description'].progress_apply(clean_text)
df3
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# Create a bag of words from the preprocessed text
vectorizer = CountVectorizer()
bag_of_words = vectorizer.fit_transform(df3['clean_text'])
```

```python
# Get a list of the words in the bag of words
words = vectorizer.get_feature_names()

# Get the number of appearances of each word
word_counts = bag_of_words.sum(axis=0)

# Create a DataFrame with the word counts
word_counts_df = pd.DataFrame({'word': words, 'count': word_counts.tolist()[0]})
```

```python
import re
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple', 'pink', 'brown', 'black', 'white', 'gray']

# Define the function to extract color from preprocessed description
def extract_color(description):
    # Search for colors in the description
    matches = re.findall('|'.join(colors), description, flags=re.IGNORECASE)
    # If no color is found, return None
    if len(matches) == 0:
        return None
    # If multiple colors are found, return the first one
    else:
        return matches[0]

# Apply the function to the 'Preprocessed Description' column to create a new 'Color' column
df3['Color'] = df3['clean_text'].progress_apply(extract_color)
```

```python
import spacy

# Load the spacy model
nlp = spacy.load('en_core_web_md')

# Function to get the Word2Vec vector for a list of words
def get_vector(full_text):
    # Initialize an empty vector
    doc = nlp(full_text)

    return doc.vector

# Apply the get_vector function to the cleaned descriptions
df3['word2vec_vector'] = df3['clean_text'].progress_apply(get_vector)
df3
# Print the resulting matrix of vectors
#vectors = np.stack(df2['word2vec_vector'].to_numpy())
#print(vectors)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# first, calculate the pairwise cosine similarity between all the vectors
similarity_matrix = cosine_similarity(np.stack(np.array(df3['word2vec_vector']),axis=0))

most_similar_indexes = []
most_similar_cosine_sims = []
for i in range(len(df3)):
    # exclude self-similarity by setting diagonal elements to -1
    similarity_matrix[i, i] = -1
    # get the indices of the 5 rows with the highest cosine similarity
    most_similar_indices = similarity_matrix[i].argsort()[:-6:-1]
    most_similar_indexes.append(np.array(df3.loc[most_similar_indices,"StockCode"]))
    # get the cosine similarity values of the 5 most similar rows
    most_similar_cosine_sim = similarity_matrix[i, most_similar_indices]
    most_similar_cosine_sims.append(most_similar_cosine_sim)

# add the most similar indexes and cosine similarity values as new columns in the dataframe
df3['most_similar_indexes'] = most_similar_indexes
df3['most_similar_cosine_sims'] = most_similar_cosine_sims
# # add the most similar indexes as a new column in the dataframe
# df['most_similar_idx'] = most_similar_indexes
df4 = df3.loc[:,['StockCode','word2vec_vector','most_similar_indexes','most_similar_cosine_sims']]

df4
```
